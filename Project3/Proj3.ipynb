```python
def eigenSolver(low, high, rows, cols):
    eigenvalues, eigenvectors = \
    LA.eig(np.random.randint(low = low, high= high, size=(rows, cols)))
    return [eigenvalues, eigenvectors]
```

```python
low = 0
high = 10
rows = 2
cols = 2
```

```python
 time = timeit.repeat(lambda: eigenSolver(low, high, rows,cols), repeat = 100, number=1)
```
